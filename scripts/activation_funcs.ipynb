```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of "obtained" values
o = np.linspace(0, 1, 500)
r = 1  # required

# Step function
f_step = np.where(o < r, 0, 1)

# Linear function
f_linear = o / r

# Logarithmic function (k = 5)
k_log = 5
f_log = np.log(1 + k_log * o) / np.log(1 + k_log * r)

# Exponential function (k = 5)
k_exp = 5
f_exp = (np.exp(k_exp * o) - 1) / (np.exp(k_exp * r) - 1)

# Polynomial functions
n_convex = 2  # convex
f_poly_convex = (o / r) ** n_convex

n_concave = 0.5  # concave
f_poly_concave = (o / r) ** n_concave

# Huber-like function
f_huber = np.where(o < r, 0.5 * (o / r) ** 2, 1)

# Plot all functions
plt.figure(figsize=(10, 6))
plt.plot(o, f_step, label="Step")
plt.plot(o, f_linear, label="Linear")
plt.plot(o, f_log, label="Logarithmic (k=5)")
plt.plot(o, f_exp, label="Exponential (k=5)")
plt.plot(o, f_poly_convex, label="Polynomial (n=2, convex)")
plt.plot(o, f_poly_concave, label="Polynomial (n=0.5, concave)")
plt.plot(o, f_huber, label="Huber-like")

plt.title("Different Incentive Functions (r = 1)")
plt.xlabel("Obtained (o)")
plt.ylabel("f(r=1, o)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
